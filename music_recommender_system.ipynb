```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfid=TfidfVectorizer(analyzer='word',stop_words='english')
```

```python
matrix=tfid.fit_transform(df['text'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similer = cosine_similarity(matrix)
print(similer)
```

Recommender Function

```python
def recommender(df, song_name):
    # Search for the title in the DataFrame
    result = df[df['song'] == song_name]  # Ensure column name 'title' is correct

    if result.empty:
        return "No recommendations available, the item wasn't found."
    else:
        # Safely return the first result if it exists
        return result.iloc[0]  # First matching result

# Example usage
response = recommender(df, 'Glow')  # Ensure 'df' contains the data
print(response)
```
